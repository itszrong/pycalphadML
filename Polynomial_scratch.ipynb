```python
def model(X, c1, c2, c3, c4, c5, c6, c7, c8, c9, c10, c11, c12, c13, c14, c15, c16, c17, c18, c19, c20, c21, c22, c23, c24, c25, c26, c27, c28, c29, c30, c31, c32, c33, c34, c35, c36, c37, c38, c39, c40,c41):
    T, XMg, XSi, XZn  = X[:,0], X[:,1], X[:,2], X[:,3]
    # print(T.shape, XMg.shape, XSi.shape, XZn.shape)

    fit = XMg*XZn*(c1+c2*T+c3*T*np.log(T)+c4*T**2)
    fit += XMg*XZn*(XMg-XZn)*(c5+c6*T)
    fit += XMg*XZn*(XMg-XZn)**2*(c7)
    fit += XMg*XSi*(c8+c9*T)
    fit += XMg*XSi*(XMg-XSi)*(c10)
    fit += XMg*XSi*(XMg-XSi)**2*(c11)
    fit += XSi*XZn*(c12)
    fit += XSi*XMg*(XSi-XMg)*(c13)
    fit += XSi*XMg*(XSi-XMg)**2*(c14)
    fit += XMg*XSi*XMg*(c15+c16*T)

    # fit = XMg*XZn*(c1+c2*T+c3*T*np.log(T)+c4*T**2)
    # fit += XMg*XZn*(XMg-XZn)*(c5+c6*T+c7*T*np.log(T)+c8*T**2)
    # fit += XMg*XZn*(XMg-XZn)**2*(c9+c10*T+c11*T*np.log(T)+c12*T**2)
    # fit += XMg*XSi*(c13+c14*T+c15*T*np.log(T)+c16*T**2)
    # fit += XMg*XSi*(XMg-XSi)*(c17+c18*T+c19*T*np.log(T)+c20*T**2)
    # fit += XMg*XSi*(XMg-XSi)**2*(c21+c22*T+c23*T*np.log(T)+c24*T**2)
    # fit += XSi*XZn*(c25+c26*T+c27*T*np.log(T)+c28*T**2)
    # fit += XSi*XMg*(XSi-XMg)*(c29+c30*T+c31*T*np.log(T)+c32*T**2)
    # fit += XSi*XMg*(XSi-XMg)**2*(c33+c34*T+c35*T*np.log(T)+c36*T**2)
    # fit += XMg*XSi*XMg*(c37+c38*T+c39*T*np.log(T)+c40*T**2)
    # fit += c41

    # fit = XMg*XZn*(c1+c2*T+0*c3*T*np.log(T)+c4*T**2)
    # fit += XMg*XZn*(XMg-XZn)*(c5+c6*T+0*c7*T*np.log(T)+c8*T**2)
    # fit += XMg*XZn*(XMg-XZn)**2*(c9+c10*T+0*c11*T*np.log(T)+c12*T**2)
    # fit += XMg*XSi*(c13+c14*T+0*c15*T*np.log(T)+c16*T**2)
    # fit += XMg*XSi*(XMg-XSi)*(c17+c18*T+0*c19*T*np.log(T)+c20*T**2)
    # fit += XMg*XSi*(XMg-XSi)**2*(c21+c22*T+0*c23*T*np.log(T)+c24*T**2)
    # fit += XSi*XZn*(c25+c26*T+0*c27*T*np.log(T)+c28*T**2)
    # fit += XSi*XMg*(XSi-XMg)*(c29+c30*T+0*c31*T*np.log(T)+c32*T**2)
    # fit += XSi*XMg*(XSi-XMg)**2*(c33+c34*T+0*c35*T*np.log(T)+c36*T**2)
    # fit += XMg*XSi*XMg*(c37+c38*T+0*c39*T*np.log(T)+c40*T**2)
    return fit
```
